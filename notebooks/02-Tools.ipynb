# The Tools Landscape

There are many tools that support working reproducibly. They combine the technologies described in the building blocks section to offer a unified package. They range in the amount they expose the underlying building blocks to end users.

## Beaker

* http://beakerx.com

## Binder

* http://mybinder.org
* https://binderhub.readthedocs.io/en/latest/

## Domino

* https://www.dominodatalab.com/product/

## DVC

* https://dvc.org
* https://iterative.ai

## Gigantum

* https://gigantum.com

## OSF

* https://cos.io/our-products/osf/

## Pachyderm

* https://www.pachyderm.io

## Stencila

* https://stenci.la

## WholeTale

* https://wholetale.org

There are two solutions we will look at in greater detail in the hands-on.

## **Renku**
* https://renkulab.io

## **CodeOcean**
* https://codeocean.com
